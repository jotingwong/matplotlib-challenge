```python
# Author: JoTing Wong

# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)

mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trail_data = pd.read_csv(clinical_trial_data_to_load)
 
# Combine the data into a single dataset

mouse_clinic = pd.merge(clinical_trail_data, mouse_drug_data, on="Mouse ID")

# Display the data table for preview

mouse_clinic.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# tumor_volume_mean = drug_timepoint_group["Tumor Volume (mm3)"].mean()
# Convert to DataFrame

drug_timepoint_group = mouse_clinic.groupby(["Drug","Timepoint"], as_index=False).agg({"Tumor Volume (mm3)": "mean"})

drug_timepoint_mean_pd = pd.DataFrame(drug_timepoint_group)

# Preview DataFrame

drug_timepoint_mean_pd.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame

std_error_group = mouse_clinic.groupby(["Drug","Timepoint"], as_index=False).agg({"Tumor Volume (mm3)": "sem"})

std_error_pd = pd.DataFrame(std_error_group)


# Preview DataFrame

std_error_pd.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mean_pivot_df = drug_timepoint_mean_pd.pivot(index="Timepoint", columns="Drug",values = "Tumor Volume (mm3)")

# Preview that Reformatting worked

mean_pivot_df.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

x_axis = np.arange(0, 50, 5)

# ax.errorbar(x_axis, means, standard_errors, fmt="o")
ax.errorbar(x_axis,mean_pivot_df["Capomulin"],yerr=True,linestyle="--",fmt="o",color="red",label="Capomulin")
ax.errorbar(x_axis,mean_pivot_df["Infubinol"],yerr=True,linestyle="--",fmt="^",color="blue",label="Infubino")
ax.errorbar(x_axis,mean_pivot_df["Ketapril"],yerr=True,linestyle="--",fmt="s",color="green",label="Ketapril")
ax.errorbar(x_axis,mean_pivot_df["Placebo"],yerr=True,linestyle="--",fmt="d",color="black",label="Placebo")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")
ax.yaxis.grid()
plt.legend(loc="upper left")

# Save the Figure

plt.savefig("Tumor Response to Treatment.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame

mean_met_group = mouse_clinic.groupby(["Drug","Timepoint"], as_index=False).agg({"Metastatic Sites": "mean"})

metastatic_df = pd.DataFrame(mean_met_group)

# Preview DataFrame
metastatic_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame

stderr_met_group = mouse_clinic.groupby(["Drug","Timepoint"],as_index=False).agg({"Metastatic Sites": "sem"})

metastatic_stderr_df = pd.DataFrame(stderr_met_group)

# Preview DataFrame
metastatic_stderr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

met_pivot_df = metastatic_df.pivot(index="Timepoint", columns="Drug",values = "Metastatic Sites")

# Preview that Reformatting worked

met_pivot_df.head()
```

```python
# Generate the Plot (with Error Bars)

fig, ax = plt.subplots()

x_axis = np.arange(0, 50, 5)

# ax.errorbar(x_axis, means, standard_errors, fmt="o")
ax.errorbar(x_axis,met_pivot_df["Capomulin"],yerr=True,linestyle="--",fmt="o",color="red",label="Capomulin")
ax.errorbar(x_axis,met_pivot_df["Infubinol"],yerr=True,linestyle="--",fmt="^",color="blue",label="Infubino")
ax.errorbar(x_axis,met_pivot_df["Ketapril"],yerr=True,linestyle="--",fmt="s",color="green",label="Ketapril")
ax.errorbar(x_axis,met_pivot_df["Placebo"],yerr=True,linestyle="--",fmt="d",color="black",label="Placebo")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (days)")
plt.ylabel("Met. Sites")
plt.ylim(0, 3.5)
ax.yaxis.grid()
plt.legend(loc="upper left")

# Save the Figure

plt.savefig("Metastatic Spread During Treatment.png")

plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame

drug_timepoint_count_group = mouse_clinic.groupby(["Drug","Timepoint"], as_index=False).agg({"Mouse ID": "count"})

mouse_count_df = pd.DataFrame(drug_timepoint_count_group)

# Preview DataFrame
mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mouse_pivot_df = mouse_count_df.pivot(index="Timepoint", columns="Drug",values = "Mouse ID")

# Preview that Reformatting worked

mouse_pivot_df


```

```python
# calculate percentage for plot
mouse_pivot_df["Capomulin_pct"] = (mouse_pivot_df["Capomulin"]/mouse_pivot_df["Capomulin"].iloc[0])*100
mouse_pivot_df["Infubinol_pct"] = (mouse_pivot_df["Infubinol"]/mouse_pivot_df["Infubinol"].iloc[0])*100
mouse_pivot_df["Ketapril_pct"] = (mouse_pivot_df["Ketapril"]/mouse_pivot_df["Ketapril"].iloc[0])*100
mouse_pivot_df["Placebo_pct"] = (mouse_pivot_df["Placebo"]/mouse_pivot_df["Placebo"].iloc[0])*100

mouse_pivot_df
```

```python
# Generate the Plot (with Error Bars)

# fig, ax = plt.subplots()

# x_axis = np.arange(0, 50, 5)



# ax.errorbar(x_axis, means, standard_errors, fmt="o")
plt.plot(x_axis,mouse_pivot_df["Capomulin_pct"],linestyle="--",marker="o",color="red",label="Capomulin")
plt.plot(x_axis,mouse_pivot_df["Infubinol_pct"],linestyle="--",marker="^",color="blue",label="Infubino")
plt.plot(x_axis,mouse_pivot_df["Ketapril_pct"],linestyle="--",marker="s",color="green",label="Ketapril")
plt.plot(x_axis,mouse_pivot_df["Placebo_pct"],linestyle="--",marker="d",color="black",label="Placebo")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
# plt.ylim(perc)
plt.grid()
plt.legend(loc="best")

# Save the Figure

plt.savefig("Survival During Treatment.png")

plt.show()
```
